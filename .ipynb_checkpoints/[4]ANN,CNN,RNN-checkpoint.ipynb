# `-` ANN 과 CNN 과 RNN

-----
`NN` 이란 neural network 로 인간의 신경망을 본 떠 만들어 선형함수와 미분을 이용하여 계산하여 우리가 찾아내지못한 알고리즘을 가중치와 편향 등으로 인공지능이 대신 찾도록해주는 딥러닝 기술의 핵심기술이다. 


<br>
----

## `1` ANN 

<br>

ANN 은 `Artificial` Neural Network 의 약자로 **`인공`** 이란 점을 강조하여 기초적인 structure 을 보여줄 때 자주 사용하는 말이다.  


[설명] <br>
input 은 input layer , hidden 은 hidden layers , output은 output layer 라고 보면된다. 
<br>
이들은 전통적인 기초 설명에 필요한 개념으로 딥러닝 이해와 활용에 매우 중요한 역할을 한다.
<br>


`0.` layer
<br>

`-` layer(레이어)는 한국어로 층이라고 불리고 그림으론 세로로 한 줄을 의미한다. 
<br>
`--` layer 내부의 동그라미는 perceptron 을 의미하며 내부에는 가중치와 편향들의 곱으로 이루어진 선형함수가 존재하고 딥러닝에서 보통 `forward()` 시에 선형함수와 비선형함수를 한 번씩 교차시키며 선형함수를 진행 후 비선형함수를 진행하는 것까지를 `1번`으로 perceptron 하나로 표현한다.
<br>
`---` 예시 : $w_{11}*x_1 + w_{21}*x_2 = 10 -> ReLU 적용 : 10 -> 10 $ , $w_{12}*x_1 + w_{22}*x_2 = -10 -> ReLU 적용 : -8 -> 8 $


`1.` input layer <br>
`-` input 데이터를 받는 곳으로 계산하기 전 모델에 맞게 정제시켜야 하는 데이터.
<br>
`2.` hidden layer <br>
`-` 중간의 계산하는 layer 층으로 개수를 많이 늘릴 수 있고 히든 레이어층이 없다면 이를 `single-layer 층` 이라고 부른다 <br>
`--` 최근 경향에 따르면 hidden layer 의 비밀이 다 밝혀져서 `hidden layer` 라 부르지 않는 추세라고 한다.<br>
`3.` output layer <br>
`-`  마지막 예측층이다. 1개에서 조금 더 많은 뉴런을 보여할 수 있고 binary(0or1) 분류이면 2개를 아웃풋으로 만들고 분류할 개수만큼 output 의 뉴런을 만든다. <br>
`--` 설명 : On the other hand,  the latter case consists of models that output the probability of an instance belonging to each of the possible class labels.

>

Convolutional layers <br>
Pooling layers <br>

<br>
다 쉬우므로 인터넷참고바람.

<br>
Fully connected layers 만 정리 진행 <br>

`Fullyconnected layers` : 마지막 layer 로 탐지에만 쓰이고 실제로 분류에 쓰이지않은 network 에 대해서 예측에 사용되도록 하는 층이다. 실제 input , output data 의 shape 에서 상당히 중요하다.

# 3 RNN

`RNN` : `recurrent neural network` 로 재사용한다는 의미를 가지고 있고 이는 vanish gradient 를 해결하기 위해 도입되었다. 
<br>
`그러나` 이를 잘 해결하지 못하자 이후 RNN 모델 내부를 수리하여 LSTM 이라는 모델을 만들어서 훨씬 잘 해결하여 상용화가 시작되었다.
<br>
RNN 모델은 `sequence` 를 다르며 연속된 데이터들을 절차에 맞게 `sequence of input(s)` , `sequence of outputs` 들을 다룬다.

<br>
`RNN` 은 일대일 , 일대다 , 다대일 , 다대다로 나타낼 수 있으며 다대다모델이 2가지로 나누어서 굳이 나누자면 5가지로 모델의 양상이 다르게 된다.
<br>
`RNN` 모델에 대해서는 꽤나 많은 공부가 필요하고 input 데이터뿐 아니라 RNN 진행방식에 대해 많은 알고리즘을 심도있게 다룰 필요가 있으므로 이번 책이 아닌 `텐초의 파이토치 딥러닝 특강` 책에서 자세하게 언급하겠습니다.

<br> <br>
간단한 RNN 모델에 대한 지식은 인터넷검색이나 유튜브 보기, 책 읽기 추천
